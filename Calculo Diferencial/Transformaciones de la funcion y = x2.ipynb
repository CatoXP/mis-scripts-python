```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(-10, 10, 400)
c = 2
def f (x):
    return x**2
# Definimos las funciones
y0= x**2              #funcin Original
y1 = x**2 + 2         # Hacia arriba
y2 = x**2 - 2         # Hacia abajo
y3 = (x + 2)**2       # Hacia la izquierda
y4 = (x - 2)**2       # Hacia la derecha
y9 = c * f(x)          # Opcin A: y = c*f(x)
y10 = (1/c) * f(x)      # Opcin B: y = (1/c)*f(x)
y11 = f(c * x)          # Opcin C: y = f(c*x)
y12 = f((1/c) * x)      # Opcin D: y = f((1/c)*x)

# Crear el grfico
#plt.figure(figsize=(10, 6))
#plt.plot(x, y0, label='y=x**2')
#plt.plot(x, y1, label='y = x**2 + 2')
#plt.plot(x, y2, label='y = x**2 - 2')
#definir funciones:

y5 = (x + 2)**2 +2
y6 = (x - 2)**2 +2
y7 = (x - 1)**2 + 1
y8 = (x + 1)**2 - 1
y9 = c * f(x)          
y10 = (1/c) * f(x)      
y11 = f(c * x)          
y12 = f((1/c) * x)      

# Crear el grfico:

plt.plot(x, y5, label=f'(x + 2)**2 +2')
plt.plot(x, y6, label=f'y = (x - 2)**2 -2')
plt.plot(x, y7, label=f'y = (x + 1) - 1')
plt.plot(x, y8, label=f'y = (x - 1) + 1 ')
plt.plot(x, y9, label=f'A) y = {c}*f(x)')
plt.plot(x, y10, label=f'B) y = (1/{c})*f(x)')
plt.plot(x, y11, label=f'C) y = f({c}*x)')
plt.plot(x, y12, label=f'D) y = f((1/{c})*x)')


# Personalizacin
plt.title('Transformaciones de la funcin y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
