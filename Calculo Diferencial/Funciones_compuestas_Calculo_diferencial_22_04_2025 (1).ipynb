```python
"""
Funciones compuestas
"""

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, simplify, sqrt, Abs, lambdify



x = symbols("x")

# Definir las funciones para los ejercicios

f1 = 3*x-5
g1 = x**2+1

# Ejercicio numero 1 f(x)=3x-5 , g(x)= x**2+1

f_g = f1.subs(x,g1) # f(g(x))= f(x**2+1)
g_f = g1.subs(x,f1) # g(f(x))= g(3x-5)

print("Ejercicio 1")
print("f(g(x)) =", simplify(f_g))
print("f(g(x)) =",simplify(g_f))
print("¿Son iguales?", simplify(f_g-g_f) == 0)
print("-"*40)

#convertir nuestras funciones simbolicas en funciones de numpy

f_gnp = lambdify(x,f_g,modules=["numpy"])
g_fnp = lambdify(x,g_f,modules=["numpy"])

# rango de valores

x_val = np.linspace(-5,5,50)

# calcular valores de las funciones compuestas
y1=f_gnp(x_val)
y2=g_fnp(x_val)

# crear la grafica
plt.figure(figsize=(10,6))
plt.plot(x_val,y1,label="f(g(x))")
plt.plot(x_val,y2,label="g(f(x))",linestyle="--")
plt.axhline(0,color="gray", linestyle=":")
plt.axvline(0,color="gray", linestyle=":")
plt.title("Comparacion entre f(g(x)) y g(f(x))")
plt.show()

```

```python
f2 = x**3
g2 = 2*x + 1

# Ejercicio numero 2 f(x)=x^3 , g(x)=2x+1

f_g = f2.subs(x, g2)  # f(g(x)) = f(2x+1)
g_f = g2.subs(x, f2)  # g(f(x)) = g(x^3)

print("Ejercicio 2")
print("f(g(x)) =", simplify(f_g))
print("g(f(x)) =", simplify(g_f))
print("¿Son iguales?", simplify(f_g - g_f) == 0)
print("-" * 40)

f_gnp = lambdify(x, f_g, modules=["numpy"])
g_fnp = lambdify(x, g_f, modules=["numpy"])

# Crear un rango de valores para x
x_val = np.linspace(-5, 5, 50)

# Calcular los valores de las funciones compuestas
y1 = f_gnp(x_val)
y2 = g_fnp(x_val)

# Crear la gráfica
plt.figure(figsize=(10,6))
plt.plot(x_val, y1, label="f(g(x))", color="blue")
plt.plot(x_val, y2, label="g(f(x))", linestyle="--", color="red")
plt.axhline(0, color="gray", linestyle=":")
plt.axvline(0, color="gray", linestyle=":")
plt.title("Comparación entre f(g(x)) y g(f(x))")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle=":", alpha=0.7)
plt.show()


```
