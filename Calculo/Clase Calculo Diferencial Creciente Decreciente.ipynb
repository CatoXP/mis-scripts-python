```python
# Aquí empieza mi código:

import numpy as np #numpy
import matplotlib.pyplot as plt #matplotlib

def f(x):
    return 2*x+1 # 2x+1


x = np.linspace(-10,10,100)
y=f(x)


# Formato a nuestra gráfica:
plt.figure(figsize=(8,6))
plt.plot(x,y,label="f(x) = 2x+1", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

def f_expo(x):
    return np.exp(x)
    
    
x = np.linspace(-5,5,100)
y= f_expo(x)

plt.figure(figsize=(8,6))
plt.plot(x,y,label="f(x) = 2x+1", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

def func_creciente(a,b,num_puntos=100,estricta=False):
    
    """
    Esta función va a verificar si mi función matemática es creciente en un intervalo
    [a,b]

    Args:
    f: Función a evaluar
    a,b: extremos del intervalo
    num_puntos: numero de puntos a evaluar

    Return: True si mi función (x) es creciente
    
    """

    x = np.linspace(a,b,num_puntos)
    y = f(x)
    if estricta:
        return np.all(np.diff(y)>0)
    else:
        return np.all(np.diff(y)>=0)
print("f(x) = 2x+1 es decreciente en [-5,5}:", func_creciente(f(x), -5,5))
```
